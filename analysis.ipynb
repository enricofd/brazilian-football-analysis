# SVD

```python
games_svd = games_duplicated.drop(['Player', 'Confronto', 'Unnamed: 0', 'Data', 'Unnamed: 0.1', '#', 'Nation', 'Pos', 'PKwon', 'PKcon'], axis=1)
fixed_value_fans = [fans.replace(",","") for fans in games_svd["Torcida"]]
games_svd["Torcida"] = fixed_value_fans
games_svd = games_svd.reset_index().drop("index", axis=1)
games_svd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Criação da pipeline de pré-processamento

imputer = SimpleImputer(strategy='median')
scaler = StandardScaler()
one_hot_encoder = OneHotEncoder(sparse=False)

num_pipeline = Pipeline([
        ('imputer', imputer),
        ('std_scaler', scaler),
    ])
    
cat_pipeline = Pipeline([
        ('cat_encoder', one_hot_encoder),
    ])


def process_data(data_input):
    
    catergorical_labels = ["Time", "Estadio", 'Time_Adversario']
    data_copy = data_input.copy()
    c = data_copy.drop(catergorical_labels, axis=1)
    
    num_attribs = list(c)
    cat_attribs = catergorical_labels
  
    full_pipeline = ColumnTransformer([
        ('num', num_pipeline, num_attribs),
        ('cat', cat_pipeline, cat_attribs),
    ])
    return full_pipeline.fit_transform(data_input)
```

```python
data_prep = process_data(games_svd)
data_prep.shape
```

```python
u, s, vh = np.linalg.svd(data_prep, full_matrices=False)
```

```python
import matplotlib.pyplot as plt
sc = np.cumsum(s)
sc = sc/(sc[-1])
plt.plot(sc)
```

```python
vh.dot(vh.transpose())
```

```python
u.transpose().dot(u)
```

```python
ut = u[:,:5]
ut = ut / np.linalg.norm(ut, axis =1).reshape(-1,1)
ut.dot(ut.transpose())
```

```python
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

# Criação da pipeline de pré-processamento

imputer = SimpleImputer(strategy='median')
scaler = StandardScaler()
one_hot_encoder = OneHotEncoder(sparse=False)

num_pipeline = Pipeline([
        ('imputer', imputer),
        ('std_scaler', scaler),
    ])

def process_data2(data_input):
    
    catergorical_labels = "Player"
    c = data_input.drop(catergorical_labels, axis=1)
    
    num_attribs = c
  
    return num_pipeline.fit_transform(num_attribs)
```

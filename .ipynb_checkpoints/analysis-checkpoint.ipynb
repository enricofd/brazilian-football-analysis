```python
#Imports
import pandas as pd
import numpy as np
import statistics
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
```

```python
#Libs
imputer = SimpleImputer(strategy='median')
scaler = StandardScaler()

num_pipeline = Pipeline([
        ('imputer', imputer),
        ('std_scaler', scaler),
    ]) #Pipeline for numerical features

def process_data(data_input):
    
    catergorical_labels = "name" 
    num_attribs = data_input.drop(catergorical_labels, axis=1) #Use only numerical features
  
    return num_pipeline.fit_transform(num_attribs)
```

```python
#Read data
data = pd.read_excel("database.xlsx")
data
```

```python
#Show columns
data.columns
```

```python
#Feature selection
players_svd = data.drop(['playerId', 'teamName', 'playedPositions', 'positionText', "value"], axis=1)
players_svd = players_svd.reset_index().drop("index", axis=1)
players_svd
```

```python
#Preprocess data
data_prep = process_data(players_svd)
data_prep
```

```python
#SVD
u, s, vh = np.linalg.svd(data_prep, full_matrices=False)
```

```python
vh.dot(vh.transpose())
u.transpose().dot(u) 
ut = u[:,:5] #Selecting 5 most "relevant" columns
ut = ut / np.linalg.norm(ut, axis =1).reshape(-1,1) #Normalizing 
products = ut.dot(ut.transpose()) #Matrix containing similarity probability
```

```python
products.shape
```

```python
# players index
index = players_svd.index
```

```python
def get_smilar_players_ordered_by_value(player_name, similar_players_number):
    player_selection_value = -1 * (similar_players_number + 1)
    player_index = index[players_svd["name"] == player_name][0]
    player_relation = products[player_index]
    similar_player_index = np.argpartition(player_relation, player_selection_value)[player_selection_value:]
    similar_player_index_no_main = similar_player_index[similar_player_index!=player_index]
    mean = statistics.mean(data.iloc[similar_player_index]["value"])
    return sorted([(player["name"], player["value"] - mean)  for idx, player in data.iloc[similar_player_index_no_main].iterrows()], key=lambda x: x[1])
```
